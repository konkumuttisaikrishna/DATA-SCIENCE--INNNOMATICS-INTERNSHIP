```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamond = diamond.drop(diamond[diamond['x']==0].index)
diamond = diamond.drop(diamond[diamond['y']==0].index)
diamond = diamond.drop(diamond[diamond['z']==0].index)
```

```python
y = diamond['price']
X = diamond[['carat','cut','color','clarity','depth','table','x','y','z']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=100)
```

```python
X_train.head()
```

```python
print(X_train.shape, y_train.shape)

print(X_test.shape, y_test.shape) 
```

```python
X_train.head()
```

```python
X_train.dtypes
```

```python
X_train_cat = X_train.select_dtypes(include=['object'])

X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head()
```

```python
X_train_cat.head()
```

```python
X_train_cat_le = pd.DataFrame(index=X_train_cat.index)

X_train_cat_le.head()
```

```python
cut_encoder = {'Fair':1,'Good':2,'Very Good':3,'Ideal':4,'Premium':5}
X_train_cat_le['cut'] = X_train_cat['cut'].apply(lambda x : cut_encoder[x])
X_train_cat_le.head()
```

```python
color_encoder = {'J':1, 'I':2, 'H':3, 'G':4, 'F':5, 'E':6, 'D':7}
X_train_cat_le['color'] = X_train_cat['color'].apply(lambda x : color_encoder[x])
X_train_cat_le.head()
```

```python
clarity_encoder = {'I1':1, 'SI2':2, 'SI1':3, 'VS2':4, 'VS1':5, 'VVS2':6, 'VVS1':7, 'IF':8}

X_train_cat_le['clarity'] = X_train_cat['clarity'].apply(lambda x : clarity_encoder[x])

X_train_cat_le.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)
X_train_num_rescaled.head()
```

```python
X_train_transformed = pd.concat([X_train_num_rescaled,X_train_cat_le], axis=1)

X_train_transformed.head()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])

X_test_cat.head()
```

```python
X_test_cat_le = pd.DataFrame(index = X_test_cat.index)

X_test_cat_le.head()
```

```python
X_test_cat_le['cut'] = X_test_cat['cut'].apply(lambda x : cut_encoder[x])

X_test_cat_le['color'] = X_test_cat['color'].apply(lambda x : color_encoder[x])

X_test_cat_le['clarity'] = X_test_cat['clarity'].apply(lambda x : clarity_encoder[x])

X_test_cat_le.head()
```

```python
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])

X_test_num.head()
```

```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = X_test_num.columns, 
                                   index = X_test_num.index)

X_test_num_rescaled.head()
```

```python
X_test_transformed = pd.concat([X_test_num_rescaled,X_test_cat_le], axis=1)

X_test_transformed.head()
```

```python
import mlflow
```

```python
mlflow.set_tracking_uri("sqlite:///mlflow.db")

mlflow.set_experiment("diamond prices Prediction")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

from sklearn import metrics
```

```python
from pickle import dump

dump(scaler, open('D:/saikrishna data science/MLOPS/MLFLOW/model/scaler.pkl', 'wb'))
```

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "krishna")
    mlflow.set_tag("algo", "KNN")
    # log the data for each run using log_param, log_metric, log_model
    mlflow.log_param("data-path", "D:/saikrishna data science/MLOPS/diamonds.csv")
    k = 5
    mlflow.log_param("n_neighbors", k)
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train_transformed, y_train)
    y_test_pred = knn_regressor.predict(X_test_transformed)
    acc = metrics.mean_squared_error(y_test, y_test_pred)
    mlflow.log_metric("mse", acc)
    mlflow.sklearn.log_model(knn_regressor, artifact_path="models")
    mlflow.log_artifact("D:/saikrishna data science/MLOPS/MLFLOW/model/scaler.pkl")
```

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "krishna")
    mlflow.set_tag("algo", "SVR")
    # log the data for each run using log_param, log_metric, log_model
    mlflow.log_param("data-path", "D:/saikrishna data science/MLOPS/diamonds.csv")
    C = 0.2
    mlflow.log_param("C", C)
    svr_regressor = SVR(C=C)
    svr_regressor.fit(X_train_transformed, y_train)
    y_test_pred = svr_regressor.predict(X_test_transformed)
    acc = metrics.mean_squared_error(y_test, y_test_pred)
    mlflow.log_metric("mse", acc)
    mlflow.sklearn.log_model(svr_regressor, artifact_path="models")
    mlflow.log_artifact("D:/saikrishna data science/MLOPS/MLFLOW/model/scaler.pkl")
```

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "krishna")
    mlflow.set_tag("algo", "decision")
    # log the data for each run using log_param, log_metric, log_model
    mlflow.log_param("data-path", "D:/saikrishna data science/MLOPS/diamonds.csv")
    depth = 4
    mlflow.log_param("max_depth", depth)
    decision_regressor = DecisionTreeRegressor(max_depth = depth)
    decision_regressor.fit(X_train_transformed, y_train)
    y_test_pred = decision_regressor.predict(X_test_transformed)
    acc = metrics.mean_squared_error(y_test, y_test_pred)
    mlflow.log_metric("mse", acc)
    mlflow.sklearn.log_model(decision_regressor, artifact_path="models")
    mlflow.log_artifact("D:/saikrishna data science/MLOPS/MLFLOW/model/scaler.pkl")
```

```python
with mlflow.start_run():
    mlflow.set_tag("dev", "krishna")
    mlflow.set_tag("algo", "linear")
    # log the data for each run using log_param, log_metric, log_model
    mlflow.log_param("data-path", "D:/saikrishna data science/MLOPS/diamonds.csv")
    #mlflow.log_param("max_depth", depth)
    linear_regressor = LinearRegression()
    linear_regressor.fit(X_train_transformed, y_train)
    y_test_pred = linear_regressor.predict(X_test_transformed)
    acc = metrics.mean_squared_error(y_test, y_test_pred)
    mlflow.log_metric("mse", acc)
    mlflow.sklearn.log_model(linear_regressor, artifact_path="models")
    mlflow.log_artifact("D:/saikrishna data science/MLOPS/MLFLOW/model/scaler.pkl")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Enabling automatic MLflow logging for scikit-learn runs
mlflow.sklearn.autolog(max_tuning_runs=None)

with mlflow.start_run():
    tuned_parameters = [{'n_neighbors':[i for i in range(1, 51)], 'p':[1, 2]}]

    clf = GridSearchCV(
        estimator=KNeighborsRegressor(), 
        param_grid=tuned_parameters, 
        scoring='r2',
        cv=5,
        return_train_score=True,
        verbose=1
    )
    clf.fit(X_train_transformed, y_train)
    
    # Disabling autologging
    mlflow.sklearn.autolog(disable=True)
```
